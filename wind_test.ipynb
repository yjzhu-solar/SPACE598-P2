```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wind_df = pd.read_csv("../src/obs/WI_H5_SWE_151205.csv",comment="#",header=None,
names=["dateobs","ne","ne_core","te","te_core","tepar","teper","tepar_core","teper_core",
"ani","ani_core","ux","uy","uz","ux_core","uy_core","uz_core"])
wind_df["dateobs"] = pd.to_datetime(wind_df["dateobs"])
wind_df = wind_df.set_index("dateobs")
wind_df_res = wind_df.resample("1H").mean()
```

Velocity in GSE

```python
wind_df_res
```

```python
fig, axes = plt.subplots(2,1,figsize=(8,6))

axes[0].scatter(wind_df_res.index,wind_df_res["ne"],s=1)
axes[1].scatter(wind_df_res.index,wind_df_res["te"],s=1)

axes[1].set_ylim(bottom=0)
```

```python
fig, axes = plt.subplots(2,1,figsize=(8,6))

axes[0].scatter(wind_df_res.index,wind_df_res["ux_core"],s=1)
axes[1].scatter(wind_df_res.index,wind_df_res["uy_core"],s=1)

axes[1].set_ylim(bottom=0)
```
